## Distribuci√≥n beta, bernoulli y el concepto de prior conjugado

En los modelos bayesianos, la **distribuci√≥n Beta** aparece con frecuencia cuando queremos modelar la **probabilidad de √©xito** en eventos binarios, es decir, situaciones en las que solo hay dos resultados posibles: √©xito o fracaso, s√≠ o no, 1 o 0.


## 1. Distribuci√≥n Beta

La distribuci√≥n **Beta** es una distribuci√≥n continua definida en el intervalo $(0,1)$, lo que la hace ideal para modelar **par√°metros de probabilidad**.

Su funci√≥n de densidad es:

$$
p(\theta) = \frac{1}{B(\alpha, \beta)}\,\theta^{\alpha - 1}(1 - \theta)^{\beta - 1}, \quad 0 < \theta < 1
$$

donde $B(\alpha,\beta)$ es la **funci√≥n Beta**, que sirve solo como constante de normalizaci√≥n.

Los par√°metros $\alpha$ y $\beta$ se interpretan como **pseudocuentas**:

- $\alpha - 1$ representa √©xitos previos (o ‚Äúcaras imaginarias‚Äù).
- $\beta - 1$ representa fracasos previos (o ‚Äúsellos imaginarios‚Äù).

Por ejemplo:
- $\mathrm{Beta}(1,1)$ es uniforme ‚Üí no hay informaci√≥n previa.
- $\mathrm{Beta}(3,1)$ concentra m√°s probabilidad hacia $\theta$ altos ‚Üí creencia de que los √©xitos son m√°s probables.
- $\mathrm{Beta}(2,5)$ concentra hacia $\theta$ bajos ‚Üí creencia de que los fracasos son m√°s probables.


**Figura 1.** Distribuciones Beta con diferentes par√°metros. Fuente: [Wikipedia](https://en.wikipedia.org/wiki/Beta_distribution)

## 2. La distribuci√≥n Bernoulli (modelo de los datos)

Cuando el resultado solo puede ser **1 (√©xito)** o **0 (fracaso)**, el modelo que inmediatamente pensaremos para los datos es la **distribuci√≥n Bernoulli**:

$$
p(x \mid \theta) = \theta^{x}(1 - \theta)^{1 - x}, \quad x \in \{0,1\}
$$

donde $\theta$ es la probabilidad de √©xito.

Por ejemplo:
- Si $x=1$ (√©xito), entonces $p(x|\theta)=\theta$.
- Si $x=0$ (fracaso), entonces $p(x|\theta)=1-\theta$.

## 3. Similitud entre Beta y Bernoulli

La estructura de ambas expresiones es **casi id√©ntica**: tienen factores de $\theta$ y $(1-\theta)$ con exponentes.

$$
\begin{align*}
\text{Bernoulli: } & \quad p(x|\theta) \propto \theta^{x}(1-\theta)^{1-x} \\
\text{Beta: } & \quad p(\theta) \propto \theta^{\alpha-1}(1-\theta)^{\beta-1}
\end{align*}
$$

Esta coincidencia **no es casualidad**: es lo que hace que la Beta sea **conjugada** de la Bernoulli.  
Significa que al aplicar Bayes, la _posterior_ tiene **la misma forma funcional** que la previa.

> **Aqu√≠ puedes revisar cu√°les son las distribuciones conjugadas m√°s comunes: [Prior conjugate.](https://en.wikipedia.org/wiki/Conjugate_prior#Table_of_conjugate_distributions)**


## 4. Prior conjugado

En el contexto bayesiano, decimos que una distribuci√≥n $p(\theta)$ es **conjugada** para una verosimilitud $p(D|\theta)$  
si la posterior $p(\theta|D)$ pertenece a la **misma familia param√©trica** que el prior.

En el caso Beta‚ÄìBernoulli:

$$
\Theta \sim \mathrm{Beta}(\alpha, \beta), \qquad X_i \mid \Theta \sim \mathrm{Bernoulli}(\Theta)
$$

entonces:

$$
\Theta \mid D \sim \mathrm{Beta}(\alpha + x, \; \beta + n - x)
$$

donde:
- $x$ = n√∫mero de √©xitos observados,
- $n-x$ = n√∫mero de fracasos observados.

Esto permite una **actualizaci√≥n muy sencilla** de la creencia sobre $\theta$ sin tener que calcular integrales complicadas.

> üí°üí°üí° **En este caso utilizamos distribuciones conocidas y estudiadas; sin embargo, como modelad@r, puedes proponer otras distribuciones previas (priors) seg√∫n el contexto del problema. Si las distribuciones que elijas no aparecen en la [tabla de conjugadas](https://en.wikipedia.org/wiki/Conjugate_prior#Table_of_conjugate_distributions), ser√° necesario realizar todos los c√°lculos manualmente.**

```{admonition} Prior conjugado
:class: tip

Un prior conjugado es una distribuci√≥n previa que, al combinarse con la verosimilitud, produce una distribuci√≥n posterior de la misma **familia param√©trica**, es decir, la _forma funcional_ de la distribuci√≥n no cambia ‚Äî solo cambian sus par√°metros.

La _forma funcional_ es la expresi√≥n algebraica que define una familia de distribuciones. En un prior conjugado, esa forma se conserva despu√©s de observar los datos.
```

## 5. Intuici√≥n de la conjugaci√≥n

La conjugaci√≥n funciona porque al multiplicar la verosimilitud y el prior,
los exponentes de $\theta$ y $(1-\theta)$ simplemente **se suman**:

$$
\begin{align*}
p(\theta|x) &\propto p(x|\theta)p(\theta) \\
&\propto \underbrace{\theta^{x}(1-\theta)^{1-x}}_{\text{Bernoulli}} \times \underbrace{\theta^{\alpha-1}(1-\theta)^{\beta-1}}_{\text{Beta}} \\
&\propto \theta^{\alpha + x - 1}(1 - \theta)^{\beta + (1 - x) - 1}
\end{align*}
$$

‚Üí Es otra Beta, con par√°metros actualizados.


## 6. Ejemplo: probabilidad de compra en una tienda en l√≠nea

Supongamos que queremos estimar la probabilidad de que un cliente **compre un producto** al visitar una tienda online.

Cada visita puede resultar en:
- **Compra (1)**
- **No compra (0)**

Por lo tanto, modelamos cada observaci√≥n como:

$$
X_i \mid \Theta \sim \mathrm{Bernoulli}(\Theta)
$$



### Creencia previa

Antes de observar datos, suponemos que la tasa de conversi√≥n promedio es del 70%.  
Expresamos esa creencia con:

$$
\Theta \sim \mathrm{Beta}(7, 3)
$$

De esta forma, la media previa es $E[\Theta] = \frac{7}{7+3} = 0.7$.



### Datos observados

Observamos 5 nuevos clientes y registramos que 4 realizaron la compra:

$$
n = 5, \quad x = 4
$$

La verosimilitud es:

$$
p(D|\theta) = \theta^{4}(1 - \theta)^{1}
$$



### Actualizaci√≥n con Bayes

La posterior es:

$$
\Theta \mid D \sim \mathrm{Beta}(\alpha + x, \beta + n - x)
$$

$$
\Theta \mid D \sim \mathrm{Beta}(7 + 4, 3 + 1) = \mathrm{Beta}(11, 4)
$$



### Interpretaci√≥n

- Media posterior:
  $$
  E[\Theta|D] = \frac{11}{11 + 4} = 0.733
  $$
- Despu√©s de ver los datos, actualizamos nuestra creencia:  
  ahora estimamos que la probabilidad de compra es aproximadamente **73%**.


## 7. Conclusi√≥n

La **distribuci√≥n Beta** es el prior natural para par√°metros de probabilidad y su forma similar a la **Bernoulli** o **Binomial** la convierte en su **prior conjugado**.

Este par (Beta‚ÄìBernoulli o Beta‚ÄìBinomial) es fundamental en estad√≠stica bayesiana porque:
- Simplifica el proceso de actualizaci√≥n.
- Tiene una interpretaci√≥n intuitiva en t√©rminos de ‚Äúpseudocuentas‚Äù.
- Permite combinar creencias previas con datos nuevos de forma coherente y transparente.

En t√©rminos generales:

$$
\text{Posterior} \propto \text{Verosimilitud} \times \text{Prior}
$$

y, gracias a la conjugaci√≥n:

$$
\mathrm{Beta}(\alpha, \beta) \;+\; x\ \text{√©xitos} \Rightarrow \mathrm{Beta}(\alpha + x,\; \beta + n - x)
$$
